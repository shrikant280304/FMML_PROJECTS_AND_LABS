### Slope & Derivatives

Creating a function for the derivative of $f(x)$ called df(x)

````python
def df(x):
    return 2 * x + 1
````

### Gradient Descent

````python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # Gradient of the function at previous_x
    new_x = previous_x - step_multiplier * gradient  # Update new_x using the gradient

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

````python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[20, 5])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)

plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)


plt.show()
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

### Example 3 - Divergence and Overflow

#### $$h(x) = x^5 - 2x^4 + 2$$

````python
## TASK-4 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Define the function h(x)
def h(x):
    return x**5 - 2*x**4 + 2

# Define the derivative h'(x)
def h_prime(x):
    return 5*x**4 - 8*x**3

# Define the second derivative h''(x)
def h_double_prime(x):
    return 20*x**3 - 24*x**2

# Find critical points by solving h'(x) = 0
critical_points = fsolve(h_prime, [0, 1.6])  # Initial guesses around 0 and 1.6

# Evaluate the second derivative at the critical points
second_derivative_values = [h_double_prime(cp) for cp in critical_points]

# Determine the nature of the critical points
nature_of_critical_points = []
for value in second_derivative_values:
    if value > 0:
        nature_of_critical_points.append("Local Minimum")
    elif value < 0:
        nature_of_critical_points.append("Local Maximum")
    else:
        nature_of_critical_points.append("Inconclusive")

# Print critical points and their nature
for cp, nature in zip(critical_points, nature_of_critical_points):
    print(f"Critical Point: x = {cp:.4f}, Nature: {nature}")

# Plotting the function
x = np.linspace(-2, 3, 400)
y = h(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='h(x) = x^5 - 2x^4 + 2', color='blue')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')

# Mark critical points on the plot
for cp in critical_points:
    plt.plot(cp, h(cp), 'ro')  # Red points for critical points
    plt.text(cp, h(cp), f'({cp:.2f}, {h(cp):.2f})', fontsize=9, ha='right')

plt.title('Plot of h(x) = x^5 - 2x^4 + 2')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.legend()
plt.grid()
plt.ylim(-10, 10)  # Adjust y-limits for better visibility
plt.show()
````

### Example 4 - Data Viz with 3D Charts

#### Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

````python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt

# Function 1: f(x, y) = 13 - x^2 - y^2 + 1
def f1(x, y):
    return 13 - x**2 - y**2 + 1

# Gradient of function 1 (partial derivatives)
def f1_grad(x, y):
    df_dx = -2*x
    df_dy = -2*y
    return np.array([df_dx, df_dy])

# Function 2: f(x, y) = 1/r + 1 where r = 3 - x^2 - y^2
def f2(x, y):
    r = 3 - x**2 - y**2
    return 1/r + 1

# Gradient of function 2 (partial derivatives)
def f2_grad(x, y):
    r = 3 - x**2 - y**2
    df_dx = 2*x / r**2
    df_dy = 2*y / r**2
    return np.array([df_dx, df_dy])

# Gradient Descent Function
def gradient_descent(grad_func, cost_func, initial_guess, multiplier, precision, max_iter):
    x, y = initial_guess
    xy_history = [(x, y)]  # Store (x, y) values
    cost_history = [cost_func(x, y)]  # Store cost (f(x, y)) values

    for i in range(max_iter):
        grad = grad_func(x, y)
        x -= multiplier * grad[0]  # Update x
        y -= multiplier * grad[1]  # Update y

        xy_history.append((x, y))
        cost_history.append(cost_func(x, y))  # Track the cost function value

        # Check if the change in x and y is below the precision threshold
        if np.linalg.norm(grad) < precision:
            break

    return xy_history, cost_history  # Return both (x, y) values and cost values

# Run gradient descent for both functions
n = 100  # Number of iterations

# Minimize function 1: f(x, y) = 13 - x^2 - y^2 + 1
initial_guess_f1 = (3, 3)  # Initial guess for function 1
low_gamma_f1, cost_f1 = gradient_descent(f1_grad, f1, initial_guess_f1, multiplier=0.1, precision=0.0001, max_iter=n)

# Minimize function 2: f(x, y) = 1/r + 1 where r = 3 - x^2 - y^2
initial_guess_f2 = (1, 1)  # Initial guess for function 2
low_gamma_f2, cost_f2 = gradient_descent(f2_grad, f2, initial_guess_f2, multiplier=0.1, precision=0.0001, max_iter=n)

# Plot the reduction in cost for both functions
plt.figure(figsize=(12, 6))

# Plot cost reduction for function 1
plt.subplot(1, 2, 1)
plt.plot(range(len(cost_f1)), cost_f1, label="f(x, y) = 13 - x^2 - y^2 + 1", color='blue')
plt.title("Cost Function Reduction for f(x, y) = 13 - x^2 - y^2 + 1")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.grid(True)
plt.legend()

# Plot cost reduction for function 2
plt.subplot(1, 2, 2)
plt.plot(range(len(cost_f2)), cost_f2, label="f(x, y) = 1/r + 1", color='green')
plt.title("Cost Function Reduction for f(x, y) = 1/r + 1")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    return np.array([theta0_slope[0], theta1_slope[0]])
    return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
````
